```python
import numpy as np
import pandas as pd
import seaborn as sns
import json
from settings import BASE_DIR
from matplotlib import pyplot

file = open(
    "{BaseDir}/Advanced_Machine_Learning_Project/ML/third_screening_results.json".format(BaseDir=BASE_DIR)
)
c_values =[]
train_score=[]
test_score=[]
l1_ratio = []
model = []
mean_fit_time=[]
mean_train_score=[]
mean_test_score=[]
tfidf_max_features=[]

data = json.load(file)
for items in data:
    model.append(items['model'])
    mean_fit_time.append(items['mean_fit_time'])
    mean_train_score.append(items['mean_train_score'])
    mean_test_score.append(items['mean_test_score'])
    train_score.append(items['mean_train_score'])
    test_score.append(items['mean_test_score'])
    tfidf_max_features.append(items['model__tfidf_max_features'])
```

```python
sns.set(rc={"figure.figsize":(10, 10)})
graph = sns.scatterplot(
    x=mean_fit_time,
    y=mean_train_score,
    size=model,
    sizes=(200,200),
    # data=df,
    hue=model,
    # style=mean_test_score
)

graph.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
tfidf_max_features = np.sort(tfidf_max_features)
pyplot.plot(tfidf_max_features, test_score, '-o', label='Test')
pyplot.plot(tfidf_max_features, train_score, '-o', label='Train')
pyplot.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
c_values =[]
train_score=[]
test_score=[]
l1_ratio = []
model = []
mean_fit_time=[]
mean_train_score=[]
mean_test_score=[]
tfidf_max_features=[]
for items in data:
    if items['mean_train_score'] < 0.74:
        model.append(items['model'])
        mean_fit_time.append(items['mean_fit_time'])
        mean_train_score.append(items['mean_train_score'])
        mean_test_score.append(items['mean_test_score'])
        train_score.append(items['mean_train_score'])
        test_score.append(items['mean_test_score'])
        tfidf_max_features.append(items['model__tfidf_max_features'])
```

```python
sns.set(rc={"figure.figsize":(10, 10)})

graph = sns.scatterplot(
    x=mean_fit_time,
    y=mean_train_score,
    size=model,
    sizes=(200,200),
    hue=model,
)

graph.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
tfidf_max_features = np.sort(tfidf_max_features)
pyplot.plot(tfidf_max_features, test_score, '-o', label='Test')
pyplot.plot(tfidf_max_features, train_score, '-o', label='Train')
pyplot.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```

```python
tol=[]
C=[]
max_iter=[]
l1_ratio=[]
train_score=[]
test_score=[]
penalty=[]

for items in data:
    if items['mean_train_score'] < 0.74:
        if items['model'] == "TFIDFLogisticRegression":
            tol.append(items['model__tol'])
            C.append(items['model__C'])
            max_iter.append(items['model__max_iter'])
            l1_ratio.append(items['model__l1_ratio'])
            train_score.append(items['mean_train_score'])
            test_score.append(items['mean_test_score'])
            penalty.append(items['model__penalty'])
```

```python
tol = np.sort(tol)
pyplot.plot(tol, test_score, '-o', label='Test')
pyplot.plot(tol, train_score, '-o', label='Train')
pyplot.legend()
pyplot.show()
```

```python
C = np.sort(C)
pyplot.plot(C, test_score, '-o', label='Test')
pyplot.plot(C, train_score, '-o', label='Train')
pyplot.legend()
pyplot.show()
```
